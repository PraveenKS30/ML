```python
import pandas as pd

data = pd.read_csv('CancerData.csv')

data.head()
```

```python
# observing dataset shape
data.shape
```

```python
# Identifying features and target variables

data_X = data.drop('diagnosis', axis=1)
data_y = data['diagnosis']
```

```python
data_X.head()
```

```python
# observing dataset columns 
data.info()
```

```python
data.describe()
```

#### This dataset has all numeral values, no object type or categorical values. Also, there is no NULL value in any of the columns.

```python
#identifying correlation among features
data_X.corr()
```

```python
from sklearn.preprocessing import StandardScaler

#As the features are on different scale, it's better to Standardize it before moving ahead..

scaler = StandardScaler()
scaled_data_X = scaler.fit_transform(data_X)
```

```python
scaled_data_X
```

```python
# As we have lots of feature and features are coorelated, it's better
# to do dimensionality reduction and to select prominent features

from sklearn.decomposition import PCA
pca = PCA().fit(scaled_data_X)
```

```python
import matplotlib.pyplot as plt
features = range(pca.n_components_)
plt.bar(features, pca.explained_variance_)
plt.xlabel('PCA features')
plt.ylabel('variance')
plt.show()
```

#### As suggested by the above plot, we see the variances are getting decreased as we move towards max 
#### feature number . Not all features are required in our model building, it's advised to pick those where
#### we see max variances. I would go with 8 features as I don't see much variances after that.

```python
dim_scaled_data_X = PCA(n_components=8).fit_transform(scaled_data_X)
```

```python
dim_scaled_data_X.shape
```

#### Now, we have reduced our features from 31 to 8 now.
#### I think our dataset is ready now for modelling.

```python
# Splitting records into training and test set

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(dim_scaled_data_X, data_y, test_size=0.2, random_state=0, stratify =data_y) 
```

```python
print('Training set shape.{}'.format(X_train.shape))
```

```python
print('Testing set shape.{}'.format(X_test.shape))
```

```python
# importing libraries for machine learning

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import accuracy_score

logreg = LogisticRegression(random_state=0)
logreg.fit(X_train, y_train)
```

```python
y_pred = logreg.predict(X_test)
```

```python
### train and test score
print("LogisticRegression accuracy score:{}").format(accuracy_score(y_test, y_pred))
```

#### Though we got a good accuracy score, it would be better to check the confusion matrix here.

```python
### confusion matrix and classification report

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test,y_pred))
```
