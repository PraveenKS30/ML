```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cols = ['party', 'infants', 'water', 'budget', 'physician', 'salvador', 'religious', 'satellite',  
        'aid', 'missile', 'immigration', 'synfuels', 'education', 'superfund', 'crime', 'duty_free_exports', 'eaa_rsa']
dataset = pd.read_csv("house-votes-84.csv")
dataset.columns = cols
dataset.head()
```

```python
dataset['infants']= dataset['infants'].apply(lambda x: 0 if x=='n' else 1)
dataset['water']= dataset['water'].apply(lambda x: 0 if x=='n' else 1)
dataset['budget']= dataset['budget'].apply(lambda x: 0 if x=='n' else 1)
dataset['physician']= dataset['physician'].apply(lambda x: 0 if x=='n' else 1)
dataset['salvador']= dataset['salvador'].apply(lambda x: 0 if x=='n' else 1)
dataset['religious']= dataset['religious'].apply(lambda x: 0 if x=='n' else 1)
dataset['satellite']= dataset['satellite'].apply(lambda x: 0 if x=='n' else 1)
dataset['aid']= dataset['aid'].apply(lambda x: 0 if x=='n' else 1)
dataset['missile']= dataset['missile'].apply(lambda x: 0 if x=='n' else 1)
dataset['immigration']= dataset['immigration'].apply(lambda x: 0 if x=='n' else 1)
dataset['synfuels']= dataset['synfuels'].apply(lambda x: 0 if x=='n' else 1)
dataset['education']= dataset['education'].apply(lambda x: 0 if x=='n' else 1)
dataset['superfund']= dataset['superfund'].apply(lambda x: 0 if x=='n' else 1)
dataset['crime']= dataset['crime'].apply(lambda x: 0 if x=='n' else 1)
dataset['duty_free_exports']= dataset['duty_free_exports'].apply(lambda x: 0 if x=='n' else 1)
dataset['eaa_rsa']= dataset['eaa_rsa'].apply(lambda x: 0 if x=='n' else 1)

dataset.head()
```

## Supervised Learning : KNeighborsClassifier

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

### Create arrays for features and response variables

```python
X = dataset.drop('party', axis =1).values
y = dataset['party'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state= 0, stratify = y)
```

```python
knn = KNeighborsClassifier()
knn.fit(X_train,y_train)
```

```python
print("Test set score: {}".format(knn.score(X_train, y_train)))
```

```python
print("Train set score: {}".format(knn.score(X_test, y_test)))
```
